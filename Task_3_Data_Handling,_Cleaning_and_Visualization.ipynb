```python
df = sns.load_dataset('iris')
print(df.head())
```

```python
print(df.shape)       # (150, 5)
print(df.columns)     # Column names
print(df.info())      # Data types and nulls
print(df.describe())  # Summary statistics
```

```python
#Data Cleaning & Manipulation
df.loc[0:2, 'sepal_length'] = np.nan
```

```python
#Check and Handle Missing Data
print(df.isnull().sum())

# Fill missing values with the mean
df['sepal_length'] = df['sepal_length'].fillna(df['sepal_length'].mean())
```

```python
#Rename Column Example

df.rename(columns={'species': 'flower_type'}, inplace=True)
```

```python
#Filter Data

# Get only the rows where petal_length > 1.5
filtered_df = df[df['petal_length'] > 1.5]
```

```python
df.head()
```

```python
#Matplotlib - Line Plot

plt.plot(df['sepal_length'])
plt.title("Sepal Length Over Samples")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length")
plt.show()
```

```python
#Seaborn - Pairplot

sns.pairplot(df, hue="flower_type")
plt.show()
```

```python
#Plotly - Interactive Bar Chart

avg_values = df.groupby('flower_type').mean().reset_index()
fig = px.bar(avg_values, x='flower_type', y='sepal_length', title="Average Sepal Length by Flower Type")
fig.show()
```
